```python
df = pd.read_csv('updated_bmi_workouts.csv')
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation Matrix', weight='bold', size=13)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
rng = np.random.RandomState(1)
BMI = 25 + 5 * rng.rand(50)  # x1
workout_days = 10 * rng.rand(50)  # x2
training_intensity = 3 * BMI + 2 * workout_days + 10 * rng.randn(50)  # y
X = np.vstack([BMI, workout_days]).T
y = training_intensity
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(BMI, workout_days, training_intensity, c='blue', label='Actual Data')
BMI_range = np.linspace(BMI.min(), BMI.max(), 10)
workout_days_range = np.linspace(workout_days.min(), workout_days.max(), 10)
BMI_grid, workout_days_grid = np.meshgrid(BMI_range, workout_days_range)
y_grid = model.predict(
    np.vstack([BMI_grid.ravel(), workout_days_grid.ravel()]).T
).reshape(BMI_grid.shape)
ax.plot_surface(BMI_grid, workout_days_grid, y_grid, alpha=0.5, color='orange', label='Regression Plane')
ax.set_xlabel('BMI')
ax.set_ylabel('Workout Days')
ax.set_zlabel('Training Intensity')
ax.set_title('Linear Regression: Training Intensity vs BMI & Workout Days')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
```
